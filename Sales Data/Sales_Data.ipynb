```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

### Merging all datasets

```python
dfs = []
```

```python
for file in files:
    df = pd.read_csv(os.path.join(r'C:\Users\44777\OneDrive\Desktop\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data', file))
    dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```

### Data Cleaning

```python
nan_df= combined_df[combined_df.isna().any(axis=1)]
nan_df

df2= combined_df.dropna(how='all')
df2
```

### Adding Month Column

```python
df2['Month']= df2['Order Date'].str[0:2]
df2
```

```python
df2= df2[df2['Order Date'].str[0:2] != 'Or']
```

```python
df2
```

```python
df2['Month']= df2['Month'].astype('int32')
```

```python
df2
```

```python
df2['Quantity Ordered']= pd.to_numeric(df2['Quantity Ordered'])
df2['Price Each']= pd.to_numeric(df2['Price Each'])
```

### Adding a Sales column

```python
df2['Sales']= df2['Quantity Ordered']* df2['Price Each']
df2
```

### 1. What's the best month for sale?

```python
results= df2.groupby('Month').sum()
results
```

```python
months= range(1,13)
plt.bar(months, results['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(months)
plt.title('Monthly Sales')
plt.show()
```

### 2. Which city had the highest sales?

```python
df2['city']= df2['Purchase Address'].apply(lambda x:x.split(',')[1] + ' ' + x.split(',')[2].split(' ')[1])
df2
```

```python
city_results= df2.groupby('city').sum()
city_results
```

```python
cities= [city for city, df in df2.groupby('city')]

plt.bar(cities, city_results['Sales'])
plt.xlabel('City Name')
plt.ylabel('Sales')
plt.xticks(cities, rotation='vertical')
plt.title('Sales by City')
plt.show()

```

### 3. Best time for sale? 

```python
df2['Order Date']= pd.to_datetime(df2['Order Date'])
```

```python
df2['Hour']=df2['Order Date'].dt.hour
df2['Minute']= df2['Order Date'].dt.minute
```

```python
df2
```

```python
hours= [hour for hour, df in df2.groupby('Hour')]

plt.plot(hours, df2.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

### 4. What products are sold the most?

```python
product_data= df2.groupby('Product')

quantity_ordered= product_data.sum()['Quantity Ordered']
```

```python
products= [product for product, df in product_data]

plt.bar(products, quantity_ordered)
plt.xlabel('City Name')
plt.ylabel('Quantity')
plt.xticks(products, rotation='vertical')
plt.title('Quantity Ordered by Products')
plt.show()
```

```python
prices= df2.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
